### 5 примеров применения методов МО в жизни и их преимущества:

1. Рекомендательные системы (например, в онлайн-кинотеатрах):
    - Применение: предложение фильмов или сериалов, которые могут заинтересовать пользователя, на основе его прошлых просмотров, оценок и предпочтений других пользователей.
    - Преимущества:
        - повышение удовлетворенности пользователя, предлагая релевантный контент;
        - снижение усилий пользователя по поиску;
        - показ контента, который пользователь, вероятно, захочет посмотреть, что может привести к покупке подписки.
2. Медицинская диагностика (например, анализ рентгеновских снимков):
    - Применение: обнаружение патологий (например, опухолей) на рентгеновских снимках, помогая врачам в постановке диагноза.
    - Преимущества:
        - быстрый анализ большого количества снимков, что экономит время врачей;
        - обнаружение тонких деталей, которые могут быть упущены человеческим глазом;
        - возможность анализа в отдаленных районах, где нет квалифицированных специалистов.
3. Автономное вождение (например, управление автомобилем):
    - Применение: распознавание дорожных знаков, разметки, пешеходов и других объектов для автоматического управления автомобилем.
    - Преимущества:
        - снижение количества аварий, вызванных человеческим фактором (усталость, невнимательность);
        - оптимизация маршрутов и расхода топлива;
        - освобождение водителя от рутинного процесса вождения.
4. Финансовое прогнозирование (например, предсказание цен на акции):
    - Применение: анализ исторических данных и рыночных трендов для прогнозирования будущих колебаний цен на акции.
    - Преимущества:
        - использование прогнозов для принятия инвестиционных решений;
        - выявление потенциально рискованных активов;
        - алгоритмическая торговля без участия человека.
5. Обработка естественного языка (например, чат-боты):
    - Применение: автоматическое общение с пользователями, ответы на вопросы, предоставление поддержки.
    - Преимущества:
        - возможность получения помощи в любое время;
        - снижение нагрузки на операторов поддержки;
        - возможность одновременного обслуживания большого количества пользователей.

### Разница между мультиклассом и мультиметкой

Мультикласс: каждый объект относится только к одному классу из множества взаимоисключающих классов. Например, фрукт может быть только яблоком, или только грушей, или только бананом.

Мультиметка: каждый объект может относиться к нескольким классам одновременно. Например, фильм может быть одновременно "комедией", "романтикой" и "драмой".

### Является ли пример с ценами на жилье из теории классификацией регрессионной проблемы? Можно ли свести регрессионную проблему к классификации?

Цены на жилье - это классическая задача регрессии, так как мы пытаемся предсказать числовую величину (цену) на основе других признаков.

Можно ли свести регрессию к классификации?
- Да, можно дискретизировать целевую переменную (цену). Например, вместо точной цены мы можем предсказывать ценовой диапазон ("низкая", "средняя", "высокая"). В таком случае это будет задача классификации.
- Однако, при таком подходе мы теряем информацию, поскольку теряется точность и конкретные значения. Дискретизация целевой переменной может быть полезна в случаях, когда важен сам класс, а не точное значение.

### Теория

1. **Базовые понятия статистики**

    - **Среднее значение:**

        Определение: сумма всех значений в наборе данных, деленная на количество этих значений.

        Использование: показывает "центр" данных, но очень чувствительно к выбросам.

    - **Медиана:**
        
        Определение: значение, которое делит набор данных пополам, когда они упорядочены.

        Использование: более устойчива к выбросам, чем среднее.

    - **Мода:**

        Определение: значение, которое встречается чаще всего в наборе данных.

        Использование: полезно для категориальных данных и идентификации наиболее распространенных значений.

    - **Дисперсия:**

        Определение: мера разброса данных относительно их среднего значения.

        Использование: показывает, насколько "разбросаны" данные.

    - **Стандартное отклонение:**

        Определение: квадратный корень из дисперсии.
        
        Использование: также показывает разброс данных, но в тех же единицах измерения, что и сами данные. Более интерпретируемо, чем дисперсия.

    - **Процентили:**

        Определение: значения, которые делят набор данных на 100 равных частей.

        Использование: показывают распределение данных и помогают идентифицировать выбросы.

        Пример: 1-й процентиль (1% самых маленьких значений), 99-й процентиль (1% самых больших значений).

2. **Распределения**

    - **Дискретное равномерное распределение:**

        Описание: все значения имеют одинаковую вероятность (например, при бросании кубика).

        Пример: выбор случайного дня недели.

    - **Распределение Бернулли:**

        Описание: два исхода (успех/неудача) с вероятностями p и 1-p (например, подбрасывание монеты).

        Пример: проверка, купит ли клиент продукт или нет.

    - **Биномиальное распределение:**

        Описание: вероятность k успехов в n независимых испытаниях Бернулли (например, количество орлов при 10 подбрасываниях монеты).

        Пример: количество кликов на объявление из 100 показов.

    - **Распределение Пуассона:**

        Описание: вероятность определенного количества событий в определенный промежуток времени (например, количество звонков в колл-центр в час).

        Пример: количество сообщений, полученных сервером в минуту.

    - **Нормальное распределение:**

        Описание: симметричное распределение вокруг среднего значения, "колоколообразная" кривая.

        Пример: распределение роста или веса людей.

    - **Экспоненциальное распределение:**

        Описание: часто описывает время между событиями в процессе Пуассона (например, время до отказа оборудования).

        Пример: время между покупками на сайте.

3. **Выбросы**

    Определение: значения, которые значительно отличаются от основной массы данных.

    Использование: выбросы могут искажать результаты анализа и могут указывать на ошибки в данных или интересные аномалии.

    Методы обнаружения:
      - Графические методы: boxplot (диаграмма ящик с усами), гистограмма.
      - Статистические методы: правило трех сигм, использование процентилей.

4. **Доверительные интервалы**

      Определение: диапазон значений, в который, вероятно, попадет истинное значение параметра (например, среднее значение) с определенной вероятностью (например, 95%).

      Использование: позволяют оценить неопределенность оценки параметра.

5. **Гистограммы и распределения:**

      - **Гистограмма:**
        
        Определение: графическое представление распределения числовых данных.

        Использование: показывает частоту значений в определенных интервалах (бинах).

      - **Распределение**: описывает, как значения распределены в наборе данных. Нормальное распределение, биномиальное, равномерное и т.д.

      - **Функция плотности ядра:**

        Определение: метод оценки плотности вероятности распределения.

        Использование: позволяет сгладить гистограмму и получить непрерывную кривую распределения, что облегчает понимание распределения данных.

6. **Диаграммы типа «ящик с усами» (Boxplots)**

      Определение: графическое представление распределения данных, показывающее медиану, квартили, межквартильный размах (IQR) и выбросы.
      
      Использование: позволяет визуально оценить разброс, медиану и выбросы в данных.
      - Коробка: содержит 50% данных, от 1-го квартиля до 3-го. Чем короче коробка, тем меньше разброс центральных данных.
      - Линия внутри коробки: медиана. Если медиана смещена к одному из концов коробки, это указывает на асимметрию распределения (скос).
      - Усы: обычно до 1.5 IQR от коробки. IQR = Q3 - Q1. Усы показывают типичный разброс данных за пределами центральной половины.
      - Точки за усами: выбросы.

### Анализ цели

```python
# Гистограмма распределения целевой переменной 'price'
plt.figure(figsize=(10, 6))
sns.histplot(df_selected['price'], bins=50, kde=True)
plt.title('Распределение цен')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

Большая часть значений сосредоточена в одном столбце в районе цены 0 и частоты около 5000. Это происходит из-за того, что в данных присутствуют очень большие выбросы, которые "сжимают" остальное распределение. В таком виде гистограмма не дает представления о реальном распределении цен, так как большинство значений сгруппированы в одном месте. Таким образом, на данном этапе распределение цены не понятно из-за влияния выбросов.

```python
# Boxplot для целевой переменной 'price'
plt.figure(figsize=(8, 6))
plt.boxplot(x=df_selected['price'])
plt.title('Boxplot цены')
plt.xlabel('Цена')
plt.show()
```

Boxplot показывает, что основная часть данных сгруппирована внизу, а также имеются выбросы в верхней части распределения, что подтверждает необходимость их обработки.

```python
# Удаление выбросов по цене
price_1st_percentile = df_selected['price'].quantile(0.01)
price_99th_percentile = df_selected['price'].quantile(0.99)

# Создание копии DataFrame
df_filtered = df_selected[(df_selected['price'] >= price_1st_percentile) & (df_selected['price'] <= price_99th_percentile)].copy() 
print(f"Размер данных после удаления выбросов по цене: {df_filtered.shape}")
```

```python
# Гистограмма распределения цены после удаления выбросов
plt.figure(figsize=(10, 6))
sns.histplot(df_filtered['price'], bins=100, kde=True)
plt.title('Гистограмма распределения цены (после удаления выбросов)')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

После удаления выбросов, гистограмма распределения цен стала более наглядной и информативной. Основная часть цен (по частоте объявлений) сосредоточена в диапазоне от 2000 до 4000 долларов. Начиная примерно с 6000 долларов и до 13000 долларов, столбцы становятся очень низкими, что указывает на небольшое количество объявлений в этом ценовом диапазоне. Распределение при этом является скошенным влево (положительная ассиметрия), то есть, большинство объявлений имеют более низкую стоимость, а объявления с более высокими ценами встречаются реже.

Этот результат отличается от гистограммы до удаления выбросов, где основная часть данных была сгруппирована в одном столбце у минимального значения, а выбросы искажали распределение. Теперь видно четкое распределение, которое показывает, что большинство объектов имеет стоимость в диапазоне от 2000 до 4000 долларов.

```python
# Извлечение признаков из 'created'
df_train['created_year'] = df_train['created'].dt.year
df_train['created_month'] = df_train['created'].dt.month
df_train['created_day'] = df_train['created'].dt.day
df_train['created_hour'] = df_train['created'].dt.hour

df_test['created_year'] = df_test['created'].dt.year
df_test['created_month'] = df_test['created'].dt.month
df_test['created_day'] = df_test['created'].dt.day
df_test['created_hour'] = df_test['created'].dt.hour
```

```python
# Создание новых признаков
df_train['bathrooms_squared'] = df_train['bathrooms']**2
df_train['bedrooms_squared'] = df_train['bedrooms']**2

df_test['bathrooms_squared'] = df_test['bathrooms']**2
df_test['bedrooms_squared'] = df_test['bedrooms']**2
```

```python
# Масштабирование числовых признаков
numerical_features = ['bathrooms', 'bedrooms', 'bathrooms_squared', 'bedrooms_squared', 'latitude', 'longitude', 'created_year', 'created_month', 'created_day', 'created_hour']

ct = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), numerical_features)
    ], remainder='drop')
```

```python
# Разделение на обучающие и тестовые данные
X_train = df_train.drop(['interest_level', 'created', 'description', 'display_address', 'features', 'photos', 'street_address', 'listing_id', 'building_id', 'manager_id', 'price'], axis=1)
y_train = df_train['price']

X_test = df_test.drop(['created', 'description', 'display_address', 'features', 'photos', 'street_address', 'listing_id', 'building_id', 'manager_id', 'price'], axis=1)
y_test = df_test['price']
```

```python
X_train_encoded = ct.fit_transform(pd.DataFrame(X_train, columns=X_train.columns))
X_test_encoded = ct.transform(pd.DataFrame(X_test, columns=X_train.columns))

# Обучение моделей
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42, max_depth=10),
    'Random Forest': RandomForestRegressor(n_estimators=100, n_jobs=-1, max_depth=10, random_state=42)
}
```

```python
results_mae = []
results_rmse = []

for i, (name, model) in enumerate(models.items()):
    print(f"Обучение модели {i+1}/{len(models)}: {name}")
    model.fit(X_train_encoded, y_train)
    y_pred_train = model.predict(X_train_encoded)
    y_pred_test = model.predict(X_test_encoded)
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mae_test = mean_absolute_error(y_test, y_pred_test)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    results_mae.append({'model': name, 'train': mae_train, 'test': mae_test})
    results_rmse.append({'model': name, 'train': rmse_train, 'test': rmse_test})
    print(f"Модель {name} обучена.")
```

```python
results_mae_df = pd.DataFrame(results_mae)
results_rmse_df = pd.DataFrame(results_rmse)

print("Результаты MAE:")
print(results_mae_df)
print("\nРезультаты RMSE:")
print(results_rmse_df)
```
